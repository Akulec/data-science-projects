```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#creating a dataset using dictionary
```

```python
#Creating your first data set using a dictionary

data = { 
         "Hours": [1, 2, 3, 4, 5, 6, 7, 8, 9],
         "Scores":[10, 20, 30, 50, 55, 65, 75, 85,95]
}
```

```python
#Convert the dataset  into a table using pandas
df = pd.DataFrame(data)
#Showing the first few rows
df.head()

```

```python
#Representing the data on a scatterplot using matplotlib
plt.scatter(df["Hours"], df["Scores"], color ="red")

#Labelling 
plt.title("Hours studied vs Scores")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.grid(True)
#Showing the plot
plt.show()
```

```python
#Preparing the data :Splitting the data into input(X) and output(y)
X = df[["Hours"]] # 2d the input must always be in 2d
y = df["Scores"] #1d that is the output

```

```python
#Creating the model
model = LinearRegression()
```

```python
#Training the empty model
model.fit(X,y)# model.fit() tells the model to learn from the X and y
```

```python
#Create the actual data as dot and draw to the predictive line for comparison
plt.scatter(X, y, color = 'blue', label = 'Actual')
plt.plot(X, model.predict(X), color = 'red', label = 'Predicted')

#Labelling
plt.title('Hours-studied vs Scores with prediction')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.legend()

#showing the graph
plt.show()
```

```python
#Now finally it's time we test our model with other data to see if it can predict
hours_studied = 9.5
prediction = model.predict([[hours_studied]])
#Using print and f-string to display the results of the prediction
print(f"If the student studied for {hours_studied} then his/her score is {prediction[0]:.2f}")
```

```python
#Evaluating our model
#Letting our model predict all outputs
predict = model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#Evaluating
mae = mean_absolute_error(y, predict)
mse = mean_squared_error(y, predict)
rsc = r2_score(y, predict)

#Displaying the evaluation using print and f-string
print(f" MEAN_ABSOLUTE_ERROR {mae :.2f} ")
print(f" MEAN_SQUARE_ERROR {mse:.2f}")
print(f"R2_SCORE {rsc:.2f}")
```
